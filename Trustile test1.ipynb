```python
test = soup1.findAll('fieldset')
for i in range(len(test)):
#     print test[i].prettify()
    print test[i].legend.span.string
```

```python
# create empty dictionary to store data for each door model
dic = {}
dic["name"] = soup1.h1.string

# There are 6 tabs in the main view showing meta-data about one door model
test = soup1.findAll('fieldset')
for i in range(len(test)):
    # First tab is Overview
    # contains images and some details
    if test[i].legend.span.string == "Overview":
        for div in test[i].findAll('div'):
            # get all types of images of this door model
            if div.get('class')[0] == "doorthumbs":
                images = {}
                for a in div.findAll('a'):
#                     print a.get('alt')
#                     print a.get('href')
                    images[a.get('alt')] = a.get('href')
                dic["images"] = images
        
            # detail information on the right of the page
            if div.get('class')[0] == "doordetails":
                details = {}
                for p in div.findAll('p'):
                    if p.get('class')[2] == "product-detail":
                        
                        category = p.span
#                         print category.contents
                        if category.string == u'Series:\xa0':
#                             print p.a.string
                            details['Series'] = p.a.string
                        if category.string == u'Application:\xa0':
#                             print p.findAll('span')[1].get_text()
                            details['Application'] = p.findAll('span')[1].get_text()
                        if category.string == u'Thickness:\xa0':
#                             print p.findAll('span')[1].get_text()
                            details['Thickness'] = p.findAll('span')[1].get_text()
                        if category.string == u'Construction:\xa0':
                            construction = []
                            for child in p.findAll('span')[1].children:
#                                 print child.string
                                construction.append(child.string)
                            details['Construction'] = construction
                        if category.string == u'Available Fire Rating:\xa0':
                            fireRating = {}
                            for img in p.findAll('span')[1].findAll('img'):
                                fireRating[img.get('alt')] = img.get('src')
#                                 print img.get('alt')
#                                 print img.get('src')
                            details['Available Fire Rating'] = fireRating
                dic["details"] = details
        
    # sticking and panel options                  
    if test[i].legend.span.string == "Profiles":
        for view in test[i].findAll('fieldset'):
            print view.legend.span.string
            # get option names
            if view.legend.span.string == "Preview Profiles":
                # sticking options
                sticking = view.find("div",attrs={"id":"sticking"})
                stickingOptions = {}
                for p in sticking.findAll("p"):
                    ab = re.search(u'.*(?=\u2009\u2013)',p.label.string).group()
                    full = re.search(u'(?<=\u2013\u2009).*',p.label.string).group()
#                     print (ab,full)
                    stickingOptions[ab] = full
                dic["Sticking Options"] = stickingOptions
                
                # panel options
                panels = view.find("div",attrs={"id":"panels"})
                panelOptions = {}
                for p in panels.findAll("p"):
                    ab = re.search(u'.*(?=\u2009\u2013)',p.label.string).group()
                    full = re.search(u'(?<=\u2013\u2009).*',p.label.string).group()
#                     print (ab,full)
                    panelOptions[ab] = full
            
            # add option images
            if view.legend.span.string == "View Profile Cross-Sections":
                # panel options
                table1 = view.findAll('table')[0]
                for td in table1.findAll('td'):
                    imageURL =  td.a.get('href')
                    label = td.span.get_text()
                    key = re.search(u'(?<=\u201c).(?=\u201d)',label).group()
                    panelOptions[key] = [panelOptions[key],imageURL]
                dic["Panel Options"] = panelOptions
                
                # sticking options
                table2 = view.findAll('table')[1]
#                 print table2
                for td in table2.findAll('td'):
                    try:
                        imageURL = td.a.get('href')
                        label = td.span.get_text()
                        key = re.search(u'(?<=\().*(?=\)\s)',label).group()
                        stickingOptions[key] = [stickingOptions[key],imageURL]
                    except:
                        pass
                dic["Sticking Options"] = stickingOptions
```
