```python
import pandas as pd
from bs4 import BeautifulSoup,NavigableString, Tag
import os
import re
import csv
import json
from urllib2 import urlopen
from urllib import urlretrieve
```

```python
testurl = "http://emtek.com/Electronic-Locks/emtouch-brass-keypad-leverset"
```

```python
soup = BeautifulSoup(urlopen(testurl).read())
body = soup.body
div = body.find('div',attrs={'id':'pageArea'})
print div.prettify()
```

```python
def get_title(div):
    title = div.find('h1').get_text()
    title = re.search('(?<=\t\t\t).*?(?=\t)',title).group()
    return title
```

```python
def get_modelImages(div):
    images = div.find('div',attrs={'id':'galleria'})
    aList = []
    for img in images.findAll('img'):
        link = img.get('src')
        imageName = re.search('(?<=/).*',link).group()
        imageName = re.search('(?<=/).*',imageName).group()
        aList.append(imageName)
#         aList.append(link)
    return aList
```

```python
def get_options(div):
    options = div.find('div',attrs={'id':'detail_options'})
    allOptions =[]
#     store options of each type into one dictionary
    for option in div.findAll('div',attrs={'class':'detail_options_row'}):
        optionType = option.find('div',attrs={'class':'options_headline'}).get_text()
#         print optionType
        aType = {}
        aType["name"] = optionType
        
#         each option itself is also a dictionary
#         all options are put in an arrary
#         which is stored as value of type dictionary
        optionsList = []
        for cell in option.findAll('div',attrs={'class':'tooltip'}):
            aOption = {}
            aOption["image"] = cell.img.get('src')
            aOption["caption"] = cell.div.get_text()
            aOption["type"] = optionType
#               optionimg = cell.img.get('src')
#               optioncaption = cell.div.get_text()
            optionsList.append(aOption)
        
        aType["value"] = optionsList

        # return an array that contains all option type dictionaries
        allOptions.append(aType)
    
    return allOptions
get_options(div)
```

```python
def get_descriptions(div):
    description = div.find('div',attrs={'id':'detail_text'})
    aArray = []
    for ddiv in description.findAll('div'):
        aSubarray = []
        for p in ddiv.findAll('p'):
            aSubarray.append(p.get_text())
    
        desc = "|".join(aSubarray)
        aArray.append(desc)
#     result = "\n".join(aArray)
    return aArray
get_descriptions(div)
```

```python
df = pd.read_csv("emtek1.csv")
df.head()
```

```python
output = open("emtek_detail.csv","wb")
writer = csv.writer(output)

for i in range(df.shape[0]):
# for i in range(0,1):
    try:
        collection = df.collections[i]
    
        url = df["details-href"][i]
        soup = BeautifulSoup(urlopen(url).read())
        body = soup.body
        div = body.find('div',attrs={'id':'pageArea'})
    
        title = get_title(div)
        title = title.encode('ascii','ignore')
    
        images = get_modelImages(div)
        imageStr = ";".join(images)
    
        descriptions = get_descriptions(div)
    
        options = get_options(div)
        opTypeCount = len(options)
        if opTypeCount == 1:
            opCount = len(options[0]["value"])
            for j in range(opCount):
                values = options[0]["value"]
#             print values[j]["caption"]
#             print values[j]["image"]
                writer.writerow([collection, title, imageStr,
                                values[j]["type"],values[j]["caption"],values[j]["image"],
                                "","","",
                                "","","",
                                descriptions[0].encode('ascii','ignore'),descriptions[1].encode('ascii','ignore'),descriptions[2].encode('ascii','ignore'),
                                url])
            
        elif opTypeCount == 2:
#         print options[0]["name"]
            opCount1 = len(options[0]["value"])
            opCount2 = len(options[1]["value"])
            for j in range(opCount1):
                values1 = options[0]["value"]
                for k in range(opCount2):
                    values2 = options[1]["value"]
            
#                 print values1[j]["caption"]
#                 print values1[j]["image"]
#                 print values2[k]["caption"]
#                 print values2[k]["image"]
#                 print "---------"
                    writer.writerow([collection, title, imageStr,
                                    values1[j]["type"],values1[j]["caption"],values1[j]["image"],
                                    values2[k]["type"],values2[k]["caption"],values2[k]["image"],
                                    "","","",
                                    descriptions[0].encode('ascii','ignore'),descriptions[1].encode('ascii','ignore'),descriptions[2].encode('ascii','ignore'),
                                    url])
    
        elif opTypeCount == 3:
            opCount1 = len(options[0]["value"])
            opCount2 = len(options[1]["value"])
            opCount3 = len(options[2]["value"])
        
            for j in range(opCount1):
                values1 = options[0]["value"]
                for m in range(opCount2):
                    values2 = options[1]["value"]
                    for n in range(opCount3):
                        values3 = options[2]["value"]
                    
                        writer.writerow([collection, title, imageStr,
                                        values1[j]["type"],values1[j]["caption"],values1[j]["image"],
                                        values2[m]["type"],values2[m]["caption"],values2[m]["image"],
                                        values3[n]["type"],values3[n]["caption"],values3[n]["image"],
                                        descriptions[0].encode('ascii','ignore'),descriptions[1].encode('ascii','ignore'),descriptions[2].encode('ascii','ignore'),
                                        url])
    
        else:
            print opTypeCount
            print title,collection,url
    
    except:
        print i
        print collection,url
        print "====="
        pass

output.close()
```
