```python
url = base_url + "/products/panel-doors/tm-series/tm9170"
soup1 = BeautifulSoup(urlopen(url).read())
print soup1.body.prettify()
                      
```

```python
test = soup1.findAll('fieldset')
for i in range(len(test)):
#     print test[i].prettify()
    print test[i].legend.span.string
```

```python
# create empty dictionary to store data for each door model
dic = {}
dic["name"] = soup1.h1.string

# There are 6 tabs in the main view showing meta-data about one door model
fields = soup1.findAll('fieldset')
for i in range(len(fields)):
    # First tab is Overview
    # contains images and some details
    if fields[i].legend.span.string == "Overview":
        for div in fields[i].findAll('div'):
            # get all types of images of this door model
            if div.get('class')[0] == "doorthumbs":
                images = []
                for a in div.findAll('a'):
#                     print a.get('alt')
#                     print a.get('href')
                    aImage = {}
                    aImage["type"] = a.get('alt')
                    aImage["url"] = a.get('href')
                    images.append(aImage)
                dic["images"] = images
        
            # detail information on the right of the page
            if div.get('class')[0] == "doordetails":
                details = {}
                for p in div.findAll('p'):
                    print p.prettify()
                    if p.get('class')[2] == "product-detail":
                        
                        category = p.span
#                         print category.contents
                        if category.string == u'Series:\xa0':
#                             print p.a.string
                            details['Series'] = p.a.string
                        if category.string == u'Application:\xa0':
                            stri = p.get_text()
                            app = re.search(u'(?<=\xa0).*',stri).group()
                            details['Application'] = app
                            
                        if category.string == u'Thickness:\xa0':
#                             print p.findAll('span')[1].get_text()
                            details['Thickness'] = p.findAll('span')[1].get_text()
                        if category.string == u'Construction:\xa0':
                            construction = []
                            for child in p.findAll('span')[1].children:
#                                 print child.string
                                construction.append(child.string)
                            details['Construction'] = construction
                        if category.string == u'Available Fire Rating:\xa0':
                            fireRating = []
                            for img in p.findAll('span')[1].findAll('img'):
                                afireRating = {}
                                afireRating["type"] = img.get('alt')
                                afireRating["url"] = img.get('src')
                                fireRating.append(afireRating)
#                                 print img.get('alt')
#                                 print img.get('src')
                            details['Available Fire Rating'] = fireRating
                dic["details"] = details
        
    # sticking and panel options                  
    if fields[i].legend.span.string == "Profiles":
        for view in fields[i].findAll('fieldset'):
            print view.legend.span.string
            # get option names
            if view.legend.span.string == "Preview Profiles":
                # sticking options
                sticking = view.find("div",attrs={"id":"sticking"})
                stickingOptions = []
                for p in sticking.findAll("p"):
                    ab = re.search(u'.*(?=\u2009\u2013)',p.label.string).group()
                    full = re.search(u'(?<=\u2013\u2009).*',p.label.string).group()
#                     print (ab,full)
                    aOption = {}
                    aOption["type"] = ab
                    aOption["name"] = full
                    stickingOptions.append(aOption)
                dic["Sticking Options"] = stickingOptions
                
                # panel options
                panels = view.find("div",attrs={"id":"panels"})
                panelOptions = []
                for p in panels.findAll("p"):
                    ab = re.search(u'.*(?=\u2009\u2013)',p.label.string).group()
                    full = re.search(u'(?<=\u2013\u2009).*',p.label.string).group()
#                     print (ab,full)
                    aOption = {}
                    aOption["type"] = ab
                    aOption["name"] = full
                    panelOptions.append(aOption)
                dic["Panel Options"] = panelOptions
            
            # add option images
            if view.legend.span.string == "View Profile Cross-Sections":
                # panel options
                table1 = view.findAll('table')[0]
                for td in table1.findAll('td'):
                    imageURL =  td.a.get('href')
                    label = td.span.get_text()
                    key = re.search(u'(?<=\u201c).(?=\u201d)',label).group()
                    for option in panelOptions:
                        if option["type"] == key:
                            option["url"] = imageURL
                dic["Panel Options"] = panelOptions
                
                # sticking options
                table2 = view.findAll('table')[1]
#                 print table2
                for td in table2.findAll('td'):
                    try:
                        imageURL = td.a.get('href')
                        label = td.span.get_text()
                        key = re.search(u'(?<=\().*(?=\)\s)',label).group()
                        for option in stickingOptions:
                            if option["type"] == key:
                                option["url"] = imageURL
                    except:
                        pass
                dic["Sticking Options"] = stickingOptions
                
    # tab3: wood species
    if fields[i].legend.span.string == "Wood Species":
        woodSpecies = []
        for li in fields[i].findAll('li'):
            wood = dict([("type",li.span.string),("url",li.a.get('href'))])
            woodSpecies.append(wood)
        dic["Wood Species"] = woodSpecies
    
    # tab4: glass options
    if fields[i].legend.span.string == "Glass":
        glassOptions = []
        for li in fields[i].findAll('li'):
            glass = dict([("type",li.span.string),("url",li.a.get('href'))])
            glassOptions.append(glass)
        dic["Glass"] = glassOptions
        
    # tab5: leather options
    if fields[i].legend.span.string == "Leather":
        leatherOptions = []
        for li in fields[i].findAll('li'):
            leather = dict([("type",li.span.string),("url",li.a.get('href'))])
            leatherOptions.append(leather)
        dic["Leather"] = leatherOptions
    
    # tab6: Metal options
    if fields[i].legend.span.string == "Metal":
        metalOptions = []
        for li in fields[i].findAll('li'):
            metal = dict([("type",li.span.string),("url",li.a.get('href'))])
            metalOptions.append(metal)
        dic["Metal"] = metalOptions    
```

```python
for key, value in dic.iteritems():
    print "%s:" % key
    if type(value) == list:
        for i in range(len(value)):
            print value[i]
    else:
        print value
```
