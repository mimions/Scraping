```python
# color choices on the right side of page
soup = BeautifulSoup(urlopen(trex2).read())
colors = soup.find("aside", attrs={"class":"column_nav"})
for img in colors.findAll('img'):
    url = img.get('data-src')
    imageName = re.search('(?<=Trex/).*',url).group()
    urlretrieve(url,imageName)
# print colors.prettify()
```
